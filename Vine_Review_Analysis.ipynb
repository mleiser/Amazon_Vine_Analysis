```python
import pandas as pd
```

```python

vine_table = pd.read_csv("vine_table.csv")
vine_table.head()
```

```python
#Step 1:
total_votes_df = vine_table[vine_table["total_votes"] >= 20]
total_votes_df.head()
```

```python
#Step 2:
percentage_df = total_votes_df[(total_votes_df["helpful_votes"]/total_votes_df["total_votes"]) >= 0.5]
percentage_df.head()
```

```python
#Step 3:
paid_reviews_df = percentage_df[percentage_df["vine"] == 'Y']
paid_reviews_df
```

```python
#Step 4:
unpaid_reviews_df = percentage_df[percentage_df["vine"] == 'N']
unpaid_reviews_df
```

```python
#Step 5 (Paid):
total_paid = len(paid_reviews_df.index)
print("Total Paid: " + str(total_paid))
total_five_paid = len(paid_reviews_df[paid_reviews_df["star_rating"] == 5].index)
print("Total Five Star Reviews Paid: " + str(total_five_paid))
percentage_five_paid = (total_five_paid / total_paid) * 100
print("Overall Percentage of Five Star Reviews Paid: " + str(percentage_five_paid))
print()
#Step 5 (Unpaid)
total_unpaid = len(unpaid_reviews_df.index)
print("Total Unpaid: " + str(total_unpaid))
total_five_unpaid = len(unpaid_reviews_df[unpaid_reviews_df["star_rating"] == 5].index)
print("Total Five Star Reviews Unpaid: " + str(total_five_unpaid))
percentage_five_unpaid = (total_five_unpaid / total_unpaid) * 100
print("Overall Percentage of Five Star Reviews Unpaid: " + str(percentage_five_unpaid))
```
